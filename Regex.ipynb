- .       - Any Character Except New Line
- \d      - Digit (0-9)
- \D      - Not a Digit (0-9)
- \w      - Word Character (a-z, A-Z, 0-9, _)
- \W      - Not a Word Character
- \s      - Whitespace (space, tab, newline)
- \S      - Not Whitespace (space, tab, newline)

- \b      - Word Boundary
- \B      - Not a Word Boundary
- ^       - Beginning of a String
- $       - End of a String

- []      - Matches Characters in brackets
- [^ ]    - Matches Characters NOT in brackets
- |       - Either Or
- ( )     - Group

Quantifiers:
- *       - 0 or More
- +       - 1 or More
- ?       - 0 or One
- {3}     - Exact Number
- {3,4}   - Range of Numbers (Minimum, Maximum)

```python
dates="""
hello
01.04.2023

2023.04.01

2023-04-01
2023-05-23
2023-06-11
2023-07-11
2023-08-11

2023/04/02

2023_04_04
2023_04_04
"""
pattern = re.compile(r"\d\d\d\d.\d\d.\d\d")
matches = pattern.finditer(dates)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"\d\d\d\d[-/]\d\d[-/]\d\d")
matches = pattern.finditer(dates)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"\d\d\d\d[-/]0[56][-/]\d\d")
matches = pattern.finditer(dates)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"\d{4}[-/]0[5-7][-/]\d{2}")
matches = pattern.finditer(dates)

#group, start, end, span

for match in matches:
  print(match)
```

```python
my_string="""
Hello World
1223
2023-05-20
Mr Simpso
Mrs Simpson
Mr. Brown
Ms Smith
Mr. T
"""
```

```python
pattern = re.compile(r"Mr\s\w+")
matches = pattern.finditer(my_string)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"Mr\.\s\w+")
matches = pattern.finditer(my_string)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"Mr\.?\s\w+")
matches = pattern.finditer(my_string)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"Mr\.?\sT")
matches = pattern.finditer(my_string)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"(Mr|Ms|Mrs)\.?\s\w+")
matches = pattern.finditer(my_string)

#group, start, end, span

for match in matches:
  print(match)
```

```python
emails="""

ashwathkaribindi@outlook.com
karibindiashwa@gmail.com
ashwanthkaribindi@gmx.de
karibindisiva@my-domain.org
"""
```

```python
pattern = re.compile(r".[@]\w+\..*")
matches = pattern.finditer(emails)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"[a-zA-Z0-9-]+@[a-zA-Z]+\.(com|de|org)")
matches = pattern.finditer(emails)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"[a-zA-Z0-9-]+@[a-zA-Z]+\.[a-zA-Z]+")
matches = pattern.finditer(emails)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"([a-zA-Z0-9-]+)@([a-zA-Z]+)\.([a-zA-Z]+)")
matches = pattern.finditer(emails)

#group, start, end, span

for match in matches:
  print(match)
```

```python
pattern = re.compile(r"([a-zA-Z0-9-]+)@([a-zA-Z]+)\.([a-zA-Z]+)")
matches = pattern.finditer(emails)

#group, start, end, span

for match in matches:
  # print(match.group(0))
  print(match.group(1))
  # print(match.group(2))
  # print(match.group(3))
```
